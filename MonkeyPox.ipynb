```python
import requests
import os
import json


# coloque o seu token aqui
bearer_token = os.environ.get("BEARER_TOKEN")

search_url = "https://api.twitter.com/2/tweets/search/all"

# Optional params: start_time,end_time,since_id,until_id,max_results,next_token,
# expansions,tweet.fields,media.fields,poll.fields,place.fields,user.fields
query_params = {'query': ' var√≠ola dos macacos OR variola dos macacos OR #varioladosmacacos','tweet.fields':'text', 
                                                                                            'tweet.fields': 'created_at',
                                                                                            'max_results':'500',
                                                                                            'start_time':'2022-09-19T23:58:21.000Z',
                                                                                            'end_time':'2022-09-25T23:59:00Z'}


def bearer_oauth(r):
    """
    Method required by bearer token authentication.
    """

    r.headers["Authorization"] = f"Bearer {bearer_token}"
    r.headers["User-Agent"] = "v2FullArchiveSearchPython"
    return r


def connect_to_endpoint(url, params):
    response = requests.request("GET", search_url, auth=bearer_oauth, params=params)
    print(response.status_code)
    if response.status_code != 200:
        raise Exception(response.status_code, response.text)
    return response.json()


def main():
    json_response = connect_to_endpoint(search_url, query_params)
    teste = json.dumps(json_response, indent=4, sort_keys=True, ensure_ascii=True)
    # teste = str(json_response)

    with open("drive/MyDrive/MestradoPPGI/topicos-ds/teste.json","a") as arq:
        arq.write(teste)
        arq.close()

    # print(teste)
    # print(json.dumps(json_response, indent=4, sort_keys=True))


if __name__ == "__main__":
    main()
```
